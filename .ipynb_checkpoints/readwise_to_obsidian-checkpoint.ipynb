```python
import pandas as pd
import os
import glob
import numpy as np

__base_path__ = '/mnt/c/Users/ryano/Downloads'
__output_path__ = "/mnt/c/Users/ryano/Google Drive/Obsidian/Big_vault/highlights_to_import"
__upload_batch_output_path__ = './uploaded_highlights.csv' 
fpath = os.path.sep.join([__base_path__,'readwise-data.csv'])
data = pd.read_csv(fpath)
data = data[data['Note'].apply(lambda x:'uploaded' not in str(x))]
#for file_ in glob.glob(__output_path__+"/*"):
#    os.remove(file_)
non_tag_keys = {'Highlight','Note','Tags','Location', 'Color','Location Type', 'Amazon Book ID'}
def parse_value(key,value):
    if key == 'Book Author':
        return ','.join(["[[" + val + "]]" for val in value.split(',')])
    elif key not in non_tag_keys:
        return "[[" + str(value) + "]]"
    else:
        return str(value)
def generate_row(key,value):
    value = parse_value(key,value)
    return ": ".join([key,value])
def to_markdown(dict_,title_key='Highlight',obs_tags=None):
    if obs_tags is None:
        obs_tags = []
    output_str = ""
    for key,item in dict_.items():
        pair = ""
        if item is not np.nan:
            pair = generate_row(key,item)
            output_str = os.linesep.join([output_str,pair])
    if len(output_str):
        output_str = os.linesep.join([output_str]+obs_tags)
        return dict_[title_key].replace(os.path.sep,'pathsep'), output_str

def row_to_markdown(row,title_key='Highlight', obs_tags=None):
    return to_markdown(row[1].to_dict(),title_key=title_key,obs_tags=obs_tags)

for fname, text in map(lambda x:row_to_markdown(x,'Highlight',['#zettelhighlight']),data.drop("Book Author",axis=1).iterrows()):
    print(fname,text)
    output_path = os.path.sep.join([__output_path__,fname+'.md'])[:259]
    print(output_path)
    with open(output_path,'w') as f:
        f.write(text)
#for fname, text in map(lambda x:row_to_markdown(x,'Book Title',['#zettelbook']),data[['Book Title','Book Author']].drop_duplicates().iterrows(),):
 #   print(fname,text)
    #break
 #   output_path = os.path.sep.join([__output_path__.replace('highlights_to_import',"Ryan's Brain"),fname+'.md'])[:259]
    #print(output_path)
  #  with open(output_path,'w') as f:
  #      f.write(text)
data['Note'] = data['Note'].apply(lambda x:str(x)+',uploaded' if str(x) !=str(np.nan) else 'uploaded')#+',uploaded'
output_data = data.rename(columns={'Highlight':'highlight','Book Title':'title','Book Author':'author','Note':'note','Location':'location'})
output_data[['highlight', 'title', 'author', 'note', 'location']].to_csv(__upload_batch_output_path__,index=False)
os.remove(fpath)
```
