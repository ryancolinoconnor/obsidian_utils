```python
non_tag_keys = {'Highlight','Note','Tags'}
```

```python
import numpy as np
```

```python
def to_markdown(dict_):
    output_str = ""
    for key,item in dict_.items():
        pair = ""
        if item is not np.nan:
            pair = ":".join([key,
            item if str(item) in non_tag_keys else "[[" + str(item) + "]]"])
            output_str = os.linesep.join([output_str,pair])
    if len(output_str):
        output_str = os.linesep.join([output_str,'#zettelhighlight'])
        return dict_['Highlight'], output_str

def row_to_markdown(row):
    return to_markdown(row[1].to_dict())
```
